```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
```

```python
## piecewise function for player 1
def f(t):
    if t < 1/3:
        return 0
    else:
        return 0.25/t**3

t = np.arange(0., 1, 0.1)

plt.plot(t, list(map(f, t)), 'b-')
plt.title("PDF for player 1")

plt.show()
```

```python
ranges_to_intergrate = [(0, 0.1), (0.1, 0.2), (0.2, 0.3), (0.3, 0.4), (0.4, 0.5), (0.5, 0.6), (0.6, 0.7), (0.7, 0.8), (0.8, 0.9), (0.9, 1)]
probability_value_p1 = []
for i in ranges_to_intergrate:
    probability_value_p1.append(quad(f, i[0], i[1])[0])

cdf_values_p1 = []
for i in range(len(probability_value_p1)):
    v = sum(probability_value_p1[:i])
    cdf_values_p1.append(v)
    
plt.plot(t, cdf_values_p1)
plt.title("CDF values for player 1")
plt.show()
plt.scatter(t, probability_value_p1)
plt.title("Probability value for each interval")
```
